```python
#Activation functons
#1.sigmoid function
import numpy as np
import matplotlib.pyplot as plt
def sigmoidfun(x):
  return 1/(1+np.exp(-x))

x=np.linspace(-10,10)
plt.plot(x,sigmoidfun(x))
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
#tanh
def tanhfun(x):
  return np.tanh(x) 

x=np.linspace(-10,10)
plt.plot(x,tanhfun(x))
plt.title('Activation Function :tanh')
plt.show()
```

```python
#ReLu
def relufun(x):
  l=[]
  for i in x:
    if i<0:
      l.append(0)
    else:
      l.append(i)
  return l



x=np.linspace(-10,10)
plt.plot(x,relufun(x))
plt.title("Activation function: ReLu")
plt.show()  
```

```python
#SoftMax
def softmax(x):
  print(x," ",np.exp(x) / np.sum(np.exp(x), axis=0))
  return  np.exp(x) / np.sum(np.exp(x), axis=0)
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))

plt.title('Activation Function :Softmax')
plt.show()
```
